To analyze the distribution of numerical data using histograms, KDE plots, and boxplots, compare distributions across groups, and identify outliers, skewness, and data spread.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

# Two sample groups
region_A = np.random.normal(55, 8, 200)
region_B = np.random.normal(65, 12, 200)

data = pd.DataFrame({
    'Region': ['A']*200 + ['B']*200,
    'Score': np.concatenate([region_A, region_B])
})

data.head()
```

```python
plt.figure(figsize=(8,5))

plt.hist(region_A, bins=20, alpha=0.5, label='Region A')
plt.hist(region_B, bins=20, alpha=0.5, label='Region B')

plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Histogram - Region A vs Region B')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(data=data, x='Score', hue='Region', fill=True)

plt.title('KDE Plot - Region A vs Region B')
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.boxplot(x='Region', y='Score', data=data)

plt.title('Boxplot - Region A vs Region B')
plt.show()
```

The distribution analysis reveals noticeable differences between Region A and Region B. Histogram and KDE plots indicate a right-skewed distribution, primarily due to higher values present in Region B. Boxplots clearly show the presence of outliers in Region B, contributing to a wider spread and greater variability. In contrast, Region A exhibits a more compact and consistent distribution with fewer extreme values. Overall, Region B demonstrates higher dispersion and skewness, while Region A shows relative stability.


```python
# Group statistics
group_stats = data.groupby('Region')['Score'].agg(['mean','median','std','skew'])
print(group_stats)
```
